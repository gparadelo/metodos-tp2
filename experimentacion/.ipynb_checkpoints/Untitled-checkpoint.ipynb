# Creamos los folds

```python
images='ImagenesCarasRed'

def writeTrainingSetFold(name, startTest,finishTest):
    file_obj  = open("./training/training-fold-" +name + ".txt", "w") 
    for j in range(1,42):    
        for i in range(1,11): 
            if i < startTest or i >= finishTest:
                file_obj.write("../"+images+"/s"+str(j)+"/"+str(i)+".pgm, "+str(j)+",\n")
    file_obj.close()

def writeTestingSetFold(name,startTest,finishTest):
    file_obj  = open("./testing/testing-fold-" +name + ".txt", "w") 
    for j in range(1,42):    
        for i in range(1,11): 
            if i >= startTest and i < finishTest:
                file_obj.write("../"+images+"/s"+str(j)+"/"+str(i)+".pgm, "+str(j)+",\n")
    file_obj.close()
```

```python
k = 5
imagesPerClass = 10
foldSize = 10/5

for i in range(1,11,2):
    j = int(i / 2)
    writeTrainingSetFold(str(j),i,i+2)
    writeTestingSetFold(str(j),i,i+2)    
```

```python
def refreshResults(i):
    filename = "./data/results-"+str(i)+".csv"
    f = open(filename,"w+")
    f.write("k;alpha;averageAccurracy;averagePrecision;averageRecall;averageF1\n")
    f.close()
    
    
for i in range(0,5):
    refreshResults(i)
```

```python
for i in range(0,5):
    for alpha in range(1,32,5):
        for k in range(1,40,3):
            training = "../experimentacion/training/training-fold-"+str(i)+".txt"
            testing = "../experimentacion/testing/testing-fold-"+str(i)+".txt"
            output = "../experimentacion/data/results-"+str(i)+".csv"
            ! ../cmake-build-debug/caralibro -i  $training -q $testing -m 1  -o $output -k $k -alpha $alpha
            print("Done k: "+str(i)+" alpha: "+str(alpha)+", neighbours:"+str(k)+"")
```

```python
frames = [ pd.read_csv("./data/results-"+str(i)+".csv", delimiter=";") for i in range(0,5) ]
result = pd.concat(frames)

frames[0]
```

```python
def plotFrame(df,idx):
    df = df[df.alpha == 1]
    ax = df.plot(x="k", logy=False)
    ax.set_ylabel("")
    ax.set_xlabel("Valor de K" )
    title = "Fold: "+str(idx)
    plt.title(title)
    plt.title(title)
    plt.show()
```

```python
for idx,frame in enumerate(frames):
    plotFrame(frame,idx)
```
